Z- score Data Normalization - IQR

```python
df6 = pd.read_csv('https://raw.githubusercontent.com/Bash-UK/TE_Sppu_DSBDA_Lab/main/Datasets/Academics.csv')
df6
```

```python
import scipy.stats as st
```

```python
st.zscore(df6['Total'])
```

```python
q1 = df6['Total'].quantile(0.25)
q2 = df6['Total'].quantile(0.50)
q3 = df6['Total'].quantile(0.75)
```

```python
print(q1, q2, q3)
```

```python
IQR = q3 - q1
IQR
```

```python
UL = q3 + (1.5 * IQR)
LL = q1 - (1.5 * IQR)
print(UL, LL)
```

z - Score

```python
outl_z = pd.DataFrame(np.where((df6['Total']>UL)))
outl_z
```

```python
outl1_z = pd.DataFrame(np.where((df6['Total']<LL)))
outl1_z
```

```python
def func():
  for i in range(0,outl_z.size):
    print(df.loc[outl_z[i]])
    print('\n')
  for i in range(0, outl1_z.size):
    print(df.loc[outl1_z[i]])
    print('\n')

func()
```

Normalization using MIN MAX

```python
def normalize(values):
  return (values - values.min())/(values.max() - values.min())
```

```python
df1 = df.copy()

df1
```

```python
df1 = df.copy()

cols = ['FE_m','SE_m','TE_m','BE_m']

df1[cols] = df1[cols].apply(normalize)
df1
```
